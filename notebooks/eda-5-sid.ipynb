```python
weather_data = data['Weather'].str.extract(r'((\d*) degrees)?,? ?(relative humidity (\d*)%)?,? ?(wind (\d*) mph)?,? ?(wind chill (\d*))?')
weather_data = weather_data[[1, 3, 5, 7]]

weather_data.columns = ['temperature', 'humidity', 'wind speed', 'wind chill']

data = pd.concat([data, weather_data], axis=1)
```
